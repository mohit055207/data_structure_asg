# theoretical questions

1.	What are data structures, and why are they important?
-> Data structures are a specific way of organizing data in a specialized format on a computer so that the information can be organized, processed, stored, and retrieved quickly and effectively. They are a means of handling information, rendering the data for easy use.


2.	Explain the difference between mutable and immutable data types with examples?
-> Mutable data types are those whose values can be changed, whereas immutable data types are ones in which the values can't be changed.
Mutable data types example : list, dictionaries and set
Immutable data types example : strings, tuples


3.	What are the main differences between lists and tuples in Python?
-> The main difference between tuples and lists is that tuples are immutable, meaning their contents cannot be changed after creation, while lists are mutable and can be modified. Tuples are most efficient than lists.


4.	 Describe how dictionaries store data.
-> Dictionaries are used to store data values in key:value pairs. A dictionary is a collection which is ordered*, changeable and do not allow duplicates. As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.


5.	Why might you use a set instead of a list in Python?
-> The key difference between a list and a set is that list can store duplicate values, whereas a set can't.


6.	What is a string in Python, and how is it different from a list?
-> A string is a sequence of characters. A list a sequence of values which can be characters, integers or even another list (referred to as a nested list).


7.	How do tuples ensure data integrity in Python?
-> Tuples are immutable to ensure that their contents remain constant throughout their lifecycle, guaranteeing data integrity and reliability. This immutability allows tuples to be used as keys in dictionaries and elements in sets, as they can be hashed.


8.	What is a hash table, and how does it relate to dictionaries in Python?
-> In general, hash tables store key-value pairs and the key is generated using a hash function. Hash tables or has maps in Python are implemented through the built-in dictionary data type. The keys of a dictionary in Python are generated by a hashing function.


9.	Can lists contain different data types in Python?
-> a list can also contain a mix of Python types including strings, floats, and booleans.


10.	Explain why strings are immutable in Python?
-> Python strings are "immutable" which means they cannot be changed after they are created.


11.	What advantages do dictionaries offer over lists for certain tasks?
-> Thus, the pros of dictionaries are:
*   They make code much easier to read if we need to generate key:value pairs. We can also do the same with a list of lists (where inner lists are pairs of "keys" and "values"), but this looks more complex and confusing.
*   We can look up a certain value in a dictionary very quickly.

12.	Describe a scenario where using a tuple would be preferable over a list?
-> If a data is needed in a program which is not supposed to be changed, then it is better to put it in 'tuples' than in 'list'.


13.	How do sets handle duplicate values in Python?
-> In Python, sets do not allow duplicate values. If you try to add a duplicate element to a set, it will be ignored because sets automatically remove duplicates.


14.	How does the “in” keyword work differently for lists and dictionaries?
-> The in keyword is used to check for membership in both lists and dictionaries, but it works differently for each:
*   **in with Lists**
*   Checks if a value exists in the list.
*   It searches sequentially, meaning it has O(n) time complexity in the worst case.
*   **in with Dictionaries**
* Checks if a key exists in the dictionary (not values).
* It is much faster (O(1) on average) because dictionaries use hashing.


15.	Can you modify the elements of a tuple? Explain why or why not.
-> No, you cannot modify the elements of a tuple.
Tuples are immutable data structures in Python, which means that once a tuple is created, its contents cannot be modified. This is because tuples are stored in a single block of memory, and changing the contents of a tuple would require moving the entire tuple to a new location in memory.
* If you try to modify a tuple, you'll get a TypeError.


16.	What is a nested dictionary, and give an example of its use case?
-> A nested dictionary is a dictionary that contains another dictionary as its value, allowing for a hierarchical organization of data.
* Example:

```
student_data = {
'John': {'age': 20, 'grade': 'A'},
'Alice': {'age': 22, 'grade': 'B'}
}
```




17.	Describe the time complexity of accessing elements in a dictionary?
-> The time complexity of accessing elements in a dictionary is O(1), also known as constant time complexity. This means that, on average, it takes the same amount of time to access an element in a dictionary, regardless of the size of the dictionary.


18.	In what situations are lists preferred over dictionaries.
-> Lists are preferred over dictionaries in the following situations:
* When data needs to be stored in a specific order.
* When data is numerical or sequential.
* When frequent insertions or deletions are required.
* When indexing is necessary (e.g., accessing elements by position).
* In general, lists are preferred when the data is homogeneous and doesn't require key-value pairs.


19.	Why are dictionaries considered unordered, and how does that affect data retrieval?
-> Dictionaries are considered unordered because the key-value pairs are stored in a hash table, which doesn't maintain a specific order. This means that:
* The order of items may change when iterating over the dictionary.
* You can't rely on a specific order when retrieving data.
- To retrieve data, you must use the key to access the corresponding value, rather than relying on the order of the items.


20.	Explain the difference between a list and a dictionary in terms of data retrieval.
-> Here's the difference:
* **List**
* Data retrieval is done using indexing (position of the element).
* Elements are accessed using their index (e.g., my_list[0]).
* Retrieval is order-based.
* **Dictionary**
* Data retrieval is done using keys.
* Elements are accessed using their key (e.g., my_dict['name']).
* Retrieval is key-based, not order-based.
